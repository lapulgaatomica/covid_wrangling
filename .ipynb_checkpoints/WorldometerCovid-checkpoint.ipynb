```python
import pandas as pd
import json
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
# df = pd.read_html('https://www.worldometers.info/coronavirus/', header=0)[0]
# df
worldo = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
soup = BeautifulSoup(worldo.text, 'lxml')
table = soup.find_all('table')
```

```python
df = pd.read_html(str(table))[0]
columns = ['Country', 'Total Cases', 'Drop1', 'Total Deaths', 'Drop2', 'Total Recovered', 'Active Cases', 'Serious / Critical Cases', 'Drop3', 'Drop4', 'Total Tests', 'Drop5']
df.columns = columns
df.drop(['Drop1', 'Drop2', 'Drop3', 'Drop4', 'Drop5'], axis=1, inplace=True)
df.drop([0, len(df.index) -1 ], inplace=True)
df.sort_values(by='Total Cases', ascending=False, inplace=True)
df.index = range(1, len(df) + 1)
df
```

```python
columns = ['Country', 'Total Cases']
world_percent = pd.DataFrame(df, columns = columns)
world_percent['World Percentage'] = world_percent['Total Cases'] / sum(world_percent['Total Cases']) * 100
more_than_1 = world_percent[world_percent['World Percentage'] >= 1]
less_than_1 = world_percent[world_percent['World Percentage'] < 1]
others = pd.DataFrame({'Country' : 'Others', 'Total Cases' : less_than_1['Total Cases'].sum(), 'World Percentage': less_than_1['World Percentage'].sum()}, index=[16])

combined = pd.concat([more_than_1, others], axis=0)
combined.set_index('Country').plot.pie(y='World Percentage', figsize=(5,5), startangle=90, legend=False)
plt.show()
#world_percent.plot.pie(y='World Percentage')
```
